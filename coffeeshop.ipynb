```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("drinkMenu.csv")
df
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python


# To use only numeric data and remove varies inorder to replace null with mode

df[['Caffeine (mg)']] = df[['Caffeine (mg)']].apply(pd.to_numeric, errors='coerce')
#print(df["Caffeine (mg)"].mean())
df['Caffeine (mg)'].fillna(df['Caffeine (mg)'].mode()[0], inplace=True)


print(df.isnull().sum())
```

```python
print(df.duplicated().sum())
```

```python
df.drop_duplicates()
```

```python
# Finding the correlation inorder to check unnecessary cols
```

```python
import seaborn as sns
plt.figure(figsize=(20,20))
corr=df.corr()
sns.heatmap(corr,annot=True)
plt.show()
```

```python
#strongly correlated is more than 0.75
```

```python
feature = []
for i in range(len(corr.columns)):
    for j in range(i):
        if abs(corr.iloc[i, j]) > 0.75:
            col = corr.columns[i]
            if col not in feature:
                feature.append(col)

feature
```

```python
df_new = df.drop(labels=feature,axis=1) 
df_new
```

```python
plt.figure(figsize=(20,20))
ax = sns.barplot(x="Beverage", y="Calories", data=df.sort_values(by=['Calories']))
s = sns.scatterplot(data=df.sort_values(by=['Calories']), x="Beverage", y="Calories")
plt.setp(s.get_xticklabels(), rotation=90)
```
